Friedmann Eq
[a'(t)/a(t)]^2 = 8*pi*G*eps(t)/(3c^2) - k*c^2/a(t)^2 
eps being epsilon, energy density 

Rearrange
a'(t)^2 = [8*pi*G*eps(t)/(3c^2)]*a(t)^2 - k*c^2

Take the derivative with respect to time
2*a'*a" = [8*pi*G*eps(t)/(3c^2)]* [eps'*a^2 + 2*eps*a*a'] 

divide through by 2*a*a'
a"/a = [4*pi*G*eps(t)/(3c^2)]*[eps'*(a/a') + 2eps]  This equation is {A} 


From the fluid eq. 
eps'(t) + [3*a'(t)/a(t)]*[eps+P] = 0

we get
eps'(t) = -[3*a'(t)/a(t)]*[eps+P]

further rearranging 
eps'(t)*[a(t)/a'(t)] = -3[eps+P]


We can plug this relation into {A}
a"/a = [4*pi*G*eps(t)/(3c^2)]*[{-3eps-3P} + 2eps]

which yeilds 
a"/a = [4*pi*G*eps(t)/(3c^2)]*[-eps - 3P]

Finnaly, we get the 3rd equation
a"/a = -[4*pi*G*eps(t)/(3c^2)]*[eps + 3P]
Thus we have shown that this equation, the acceleartion of the universe, is derived from the Friedman and Fluid Equation

Both a & b concern a flat universe, thus (1-Omega0) goes to zero. 

```python
from scipy.integrate import quad
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

For proper distance we have the integral 
r = Integral[(c*dt)/a] = Integral{0 to z}[(c*dz)/H(z)]

We are given that 
(H(z)/H0)^2 = Omega{r}(1+z)^4 + Omega{m}(1+z)^3 + Omega{lambda} + (1-Omega{o}(1+z)^2 

We know that the universe is flat for this problem, so the Omega for curviture goes to 0, and in problem a Omega of radiation is very small, so we can have that go to zero.
Solving for H(z) we get 
H(z) = H0*sqrt[Omega{m}(1+z)^3 + Omega{lambda}]

Plugging into the integral for comoving distance we get 

r = integral{0 to z}[(c*dz)/(H0*sqrt[Omega{m}(1+z)^3 + Omega{lambda}])] 

For a we are given that 
Omega{m} = 0.3
Omega{lambda} = 0.7

```python
H0= 2.2683e-18 #seconds
H0gyr= 0.07154 #gigayears
c = 30000000 #m*s^-1

def integrand(z):
    return (c/(H0*(0.3*(1+z)**3 + 0.7)**(1/2)))

step = 0.01 
values = [] 
a = np.linspace(0,10,100) #redshift runs from 0 to 10 redshift
for i in a: 
    intergral0,err = quad(integrand,0,i)
    
    values.append(intergral0)
    
values=np.asarray(values)*3.24e-23 #convert meters to Mpc


print (values)
plt.plot(a,values)
plt.ylabel('Distance [Mpc]')
plt.xlabel('Redshift [Z]')
```

Then to solve for time, we can use the integral
Integral{0 to z}[(c*dz)/H(z)]

rearranging the integral with dz/dt = H(z)/a(t) we can get 
{Delta}*t= Integral{0 to z} [dz/((1+z)*H0*sqrt[Omega{m}(1+z)^3 + Omega{lambda}]]

{Delta}t can also be thought of as the lookback age


```python
def timefunc1(z):
    return 1/((1+z)*(H0gyr*(0.3*(1+z)**3 + 0.7)**(1/2)))

step = 0.01 
valuest1 = [] 
a = np.linspace(0,10,500)
for i in a:
    intergral1,err = quad(timefunc1,0,i)
    valuest1.append(intergral1)
    
valuest1=np.asarray(valuest1)

plt.plot(a,valuest1)

print (valuest1)
plt.ylabel('Look Back Age [Gyr]')
plt.xlabel('Redshift [Z]')
```

To find the age of the benchmark universe, we can intgrate the age of the universe over a large redshift which gives an approximate age of 13.4596 Gigeyears


```python
def timefunc1(z):
    return 1/((1+z)*(H0gyr*(0.3*(1+z)**3 + 0.7)**(1/2)))

step = 0.01 
valuest1 = [] 
a = np.linspace(0,100,500)
for i in a:
    intergral1,err = quad(timefunc1,0,i)
    valuest1.append(intergral1)
    
valuest1=np.asarray(valuest1)

print (valuest1)
```

For this matter only universe, we do not know that curviture is 0, therefore Omega{o} = Omega{m}

that means that 
H(z) = H0*sqrt[Omega{m}(1+z)^2 + (1-Omega)(1+z)^2] 

```python
massdensity = 0.01905
def timefuncm(z):
    return 1/((1+z)*(H0gyr*(massdensity*(1+z)**3 + (1-massdensity)*(1+z)**2)**(1/2)))

step = 0.01 
valuestm = [] 
a = np.linspace(0,100,500)
for i in a:
    intergralm,err = quad(timefuncm,0,i)
    valuestm.append(intergralm)
    
valuestm=np.asarray(valuestm)

print (valuestm)
```

By varrying the matter density parameter, we can find a close age of the universe for the given matter density parameter

We find that the matter density= **0.01905** is the optimal mass density

As shown above, the age of this matter only universe is **13.45977 Gyr**
